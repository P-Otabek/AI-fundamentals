```python
#Problem: which two are equivalent
#we equated a implies b to not a or b
#A. (p∧q)→r = ¬(p∧q)∨r = ¬p∨¬q∨¬r (Morgan's law)
#B. (p→q)∧(p→r) = (¬p∨q)∧(¬p∨r)
#C. p→(q∧r) = ¬p∨(q∧r) = (¬p∨q)∧(¬p∨r) (Distributive law)
# As can be seen B and C are eqivalent and below is the python code that checks the same

#function for imply
def imply(a,b):
    return not a or b
#propositions for p, q, r
p_props = [True, True, True, True, False, False, False, False]
q_props = [True, True, False, False, True, True, False, False]
r_props = [True, False, True, False, True, False, True, False]

#truth table for each proposition
A_table = []
B_table = []
C_table = []

for i in range(len(p_props)):
    p = p_props[i]
    q = q_props[i]
    r = r_props[i]
    
    #values for A
    a = imply(p and q, r)
    A_table.append((p, q, r, a))
    
    #values for B
    b = (imply(p, q) and imply(p, r))
    B_table.append((p, q, r, b))
    
    #values for A
    c = imply(p, q and r)
    C_table.append((p, q, r, c))
    
#truth table for A, B and C
print("Truth table for A:")
for row in A_table:
    print(row)
    
for row in B_table:
    print(row)

for row in C_table:
    print(row)
    
#find the equivalent statements
if A_table == B_table:
    print("A and B statements are equivalent.")
elif A_table == C_table:
    print("A and C statements are equivalent.")
elif B_table == C_table:
    print("B and C statements are equivalent.")
else:
    print("None of the statements are equivalent.")
    
```
