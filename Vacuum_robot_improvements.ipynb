```python
#ChatGPT and Google were used for most of the syntax
def clean_room(location, status, goal_state, total_cost):
    if status == '1':
        print(f"Location {location} is Dirty.")
        # suck the dirt  and mark it as clean
        goal_state[location] = '0'
        total_cost += 1  # cost for suck
        print(f"Cost for CLEANING {location}: {total_cost}")
        print(f"Location {location} has been cleaned.")

    else:
        print(f"Location {location} is already clean.")

    return total_cost, goal_state


def vacuum_world():
    # initializing goal_state
    # 0 indicates Clean and 1 indicates Dirty
    goal_state = {'A': '0', 'B': '0'}
    total_cost = 0
    
    while True:
        location_input = input("Enter Location of Vacuum (A/B): ").upper()
        if location_input not in ['A', 'B']:
            print("Invalid location. Please enter A or B.")
            continue
        status_input = input(f"Enter status of {location_input} (0 for CLEAN, 1 for DIRTY): ")
        if status_input not in ['0', '1']:
            print("Invalid status. Please enter 0 for CLEAN or 1 for DIRTY.")
            continue 
        location_input_complement = 'B' if location_input == 'A' else 'A'
        status_input_complement = input(f"Enter status of {location_input_complement} (0 for CLEAN, 1 for DIRTY): ")
        if status_input_complement not in ['0', '1']:
            print("Invalid status. Please enter 0 for CLEAN or 1 for DIRTY.")
            continue
        break
        
    print(f"Initial Location Condition: {goal_state}")

    if location_input == 'A':
        # Location A is Dirty.
        print("Vacuum is placed in Location A")
        total_cost, goal_state = clean_room('A', status_input, goal_state, total_cost)

        if status_input_complement == '1':
            # if B is Dirty
            print("Location B is Dirty.")
            print("Moving right to the Location B.")
            total_cost += 1  # cost for moving right
            print(f"COST for moving RIGHT: {total_cost}")
            total_cost, goal_state = clean_room('B', '1', goal_state, total_cost)

        else:
            print("No action")
            # suck and mark clean
            print("Location B is already clean.")

    else:
        print("Vacuum is placed in location B")
        # Location B is Dirty.
        total_cost, goal_state = clean_room('B', status_input, goal_state, total_cost)

        if status_input_complement == '1':
            # if A is Dirty
            print("Location A is Dirty.")
            print("Moving LEFT to the Location A.")
            total_cost += 1  # cost for moving left
            print(f"COST for moving LEFT: {total_cost}")
            total_cost, goal_state = clean_room('A', '1', goal_state, total_cost)

        else:
            # suck and mark clean
            print("Location A is already clean.")

    # done cleaning
    print("GOAL STATE: ")
    print(goal_state)
    print("Performance Measurement: " + str(total_cost))


vacuum_world()
```
